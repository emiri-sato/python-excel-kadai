```python
import openpyxl
from datetime import datetime

workbook=openpyxl.Workbook()
sheet=workbook.create_sheet("請求書")



sheet["B2"] = "請求書"
sheet["B4"] = "株式会社ABC"
sheet["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
sheet["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
sheet["B7"] = "担当者名:鈴木一郎 様"
sheet["F4"] = "No."
sheet["G4"] = "0001"
sheet["F5"] = "日付"
sheet["G5"]=datetime.today().strftime("%Y/%m/%d")

header=["商品名","数量","単価","金額"]
for col, item in enumerate(header, start=2):
    sheet.cell(row=10, column=col, value=item)

data=[
    ["商品A",2,10000,20000],
    ["商品B",1,15000,15000]
]

start_row=11
for i, data in enumerate(data):
    row = start_row + i
    for col, val in enumerate(data, start=2):
        sheet.cell(row=row, column=col, value=val)

sheet["E13"]=35000
sheet["B15"]="合計"
sheet["B16"]="消費税"
sheet["B17"]="税込合計"

sheet["E15"]=35000
sheet["E16"]=3500
sheet["E17"]=38500

file_name = f"請求書_{datetime.today().strftime('%Y%m%d')}.xlsx"
workbook.save(file_name)
```
